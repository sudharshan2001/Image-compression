```python
import os
import cv2 as cv
import matplotlib.pyplot as plt
import PIL,sys
import numpy as np
```

```python
img = PIL.Image.open('Images/sample6.jpg')

plt.imshow(img)
```

```python
img_1 = cv.imread("Images/sample6.jpg", cv.IMREAD_COLOR)
```

```python
r = np.linalg.svd(img_1[:,:,0])
g = np.linalg.svd(img_1[:,:,1])
b = np.linalg.svd(img_1[:,:,2])
```

```python
# try out varous values for it
i=6000
reconst_r = np.matrix(r[0][:, :i]) * np.diag(r[1][:i]) * np.matrix(r[2][:i,:])  
reconst_g = np.matrix(g[0][:, :i]) * np.diag(g[1][:i]) * np.matrix(g[2][:i,:])
reconst_b = np.matrix(b[0][:, :i]) * np.diag(b[1][:i]) * np.matrix(b[2][:i,:])
```

```python
svd_rgb = np.dstack((np.asarray(reconst_b), np.asarray(reconst_g), np.asarray(reconst_r))).astype('uint8')
plt.imshow(svd_rgb)
title = " Image after =  %s" %i
plt.title(title)
plt.show()
```

```python
PIL.Image.fromarray(svd_rgb).save('dummysample6.jpg')
```
